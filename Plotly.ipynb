# Boxplot

```python
import plotly.express as px
```

```python
df = px.data.tips()
```

```python
df
```

```python
fig = px.box(df,y='total_bill')
fig.show()
```

```python
fig = px.box(df,y='tip',title= 'box plot of tip')
fig.show()
```

```python
fig = px.box(df,x='time',y='total_bill',title= 'box plot of time vs total_bill')
fig.show()
```

```python
fig = px.box(df,x='time',y='total_bill',title= 'box plot of time vs total_bill',points='all')
fig.show()
```

```python
fig = px.box(df,x='time',y='total_bill',title= 'box plot of total_bill',color='smoker',notched=True,hover_data=['day'])
fig.show()
```

```python
fig = px.strip(df,x='day',y='tip')
fig.show()
```

# Violin plot

```python
import plotly.express as px
```

```python
df = px.data.tips()
```

```python
df.head()
```

```python
fig = px.violin(df,y='total_bill')
fig.show()
```

```python
fig = px.violin(df,y='total_bill',box=True,points='outliers')
fig.show()
```

```python
fig = px.violin(df,y='total_bill',box=True,points='all')
fig.show()
```

```python
df.columns
```

```python
fig = px.violin(df,y='tip',color='time',violinmode='overlay',hover_data=df.columns,box=True)
fig.show()
```

```python
fig = px.strip(df,x='day',y='tip')
fig.show()
```

```python
fig = px.violin(df,y='total_bill')
fig.update_traces(quartilemethod='exclusive')
fig.show()
```

```python
fig = px.violin(df,y='total_bill')
fig.update_traces(quartilemethod='inclusive')
fig.show()
```
